```python
import numpy as np
import matplotlib.pyplot as plt

# Input data
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(square_footage, selling_prices, color='green')

# Add labels and title
plt.title('Housing Prices vs. Square Footage', fontsize=14)
plt.xlabel('Square Footage (sq. ft.)', fontsize=12)
plt.ylabel('Selling Price (in $1000)', fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0'])

# Add title
plt.title('Distribution of Monthly Income by Source', fontsize=14)

# Display the chart
plt.show()
```
