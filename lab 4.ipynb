```python
import matplotlib.pyplot as plt

# Input data
days = list(range(1, 32))  # Days of the month
temperature = [65, 68, 70, 72, 75, 76, 78, 80, 81, 79, 75, 72, 70, 68, 67, 69, 70, 73, 75, 76, 78, 80, 81, 82, 83, 82, 80, 78, 76, 74, 71]

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(days, temperature, marker='o', linestyle='-', color='b')

# Add labels, title, and grid
plt.title('Daily Temperature Changes in July', fontsize=14)
plt.xlabel('Day of the Month', fontsize=12)
plt.ylabel('Temperature (°F)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data
days = list(range(1, 78))  # Days from 1 to 77
stock_prices = [
    100, 105, 110, 115, 112, 120, 118, 125, 128, 130, 132, 135, 138, 140, 142, 
    144, 145, 148, 150, 155, 160, 158, 162, 165, 170, 172, 175, 178, 180, 182, 
    185, 188, 190, 192, 195, 198, 200, 198, 195, 193, 190, 188, 185, 182, 180, 
    178, 175, 172, 170, 168, 165, 162, 160, 158, 155, 152, 150, 148, 145, 143, 
    140, 138, 135, 132, 130, 128, 125, 123, 120, 118, 115, 112, 110, 108, 105, 103, 100
]

# Create the line plot
plt.figure(figsize=(10, 6))  # Set figure size for better readability
plt.plot(days, stock_prices, marker='o', color='b', label='Stock Price')  # Line plot with blue color and markers

# Add labels and title
plt.xlabel('Day')  # Label for x-axis
plt.ylabel('Closing Price ($)')  # Label for y-axis
plt.title('Stock Daily Closing Prices Over 77 Days')  # Title of the plot

# Add a grid for better readability
plt.grid(True)

# Show the plot
plt.legend()  # Show legend
plt.show()
```
